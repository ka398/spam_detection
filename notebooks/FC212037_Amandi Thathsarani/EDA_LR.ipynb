```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/data_set.csv', encoding='latin-1')
```

```python
df = df.T
```

```python
df.head()
```

```python
df = df.T  # Transpose the DataFrame so emails are rows and features are columns
```

```python
print(df.columns)
print(df.index)
df.head()

```

```python
# 1. Drop the 'Email No.' column (not needed for training)
df = df.drop(columns='Email No.')

# 2. Convert all values to numeric (if not already)
df = df.apply(pd.to_numeric)

# 3. Separate features and target label
X = df.drop(columns='Prediction')  # Features: word counts
y = df['Prediction']               # Labels: 0 = ham, 1 = spam
```

```python
sns.countplot(x=y)
plt.title('Spam vs Ham Count')
plt.xlabel('Label (0 = Ham, 1 = Spam)')
plt.show()
```
